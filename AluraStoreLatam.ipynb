### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
# Sumando el proceso de facturacion por tienda
facturacion_total_1 = tienda1['Precio'].sum()
facturacion_total_2 = tienda2['Precio'].sum()
facturacion_total_3 = tienda3['Precio'].sum()
facturacion_total_4 = tienda4['Precio'].sum()

# Mostrando el total de facturacion por tienda
print(f'La tienda 1 tiene una facturacion total de: ${facturacion_total_1}')
print(f'La tienda 2 tiene una facturacion total de: ${facturacion_total_2}')
print(f'La tienda 3 tiene una facturacion total de: ${facturacion_total_3}')
print(f'La tienda 4 tiene una facturacion total de: ${facturacion_total_4}')


# Mostando el total de facturacion global de las tiendas
facturacion_total_global = facturacion_total_1 + facturacion_total_2 + facturacion_total_3 + facturacion_total_4
print(f'\nLa facturacion global de las tiendas es de: ${facturacion_total_global}')
```

##Desarrollo

Se muestra que claramente que la tienda 4 presenta menor facturación.


# 2. Ventas por categoría

```python
# Función para calcular y mostrar las ventas por categoría de una tienda
def ventas_por_categoria(tienda, numero):
    ventas = tienda.groupby('Categoría del Producto')['Precio'].sum()
    print(f"\nVentas por categoría en Tienda {numero}:")
    print(ventas)

# Mostrar resultados para cada tienda
ventas_por_categoria(tienda1, 1)
ventas_por_categoria(tienda2, 2)
ventas_por_categoria(tienda3, 3)
ventas_por_categoria(tienda4, 4)

#venta por Categoria por numero de ventas
venta_prod_tienda1 = tienda1['Categoría del Producto'].value_counts()
venta_prod_tienda2 = tienda2['Categoría del Producto'].value_counts()
venta_prod_tienda3 = tienda3['Categoría del Producto'].value_counts()
venta_prod_tienda4 = tienda4['Categoría del Producto'].value_counts()

print("\nVentas por categoría en la tienda 1:")
print(venta_prod_tienda1)
print("\nVentas por categoría en la tienda 2:")
print(venta_prod_tienda2)
print("\nVentas por categoría en la tienda 3:")
print(venta_prod_tienda3)
print("\nVentas por categoría en la tienda 4:")
print(venta_prod_tienda4)


```

## Desarrollo

En este punto se pueden ver que las cuatro tiendas tienen mejores ventas en la categoria muebles, seguida por la categoria electronicos.

De igual forma pero desde menor a mayor, se encuentan las categorias de articulos para el hogar, libros e instrumentos musicales, siendo estas, categorias con peores ventas.

# 4. Productos más y menos vendidos

```python
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

destacados = []

for tienda_df, nombre in zip(tiendas, nombres):
    ventas = tienda_df['Producto'].value_counts()
    mas_vendido = ventas.idxmax()
    menos_vendido = ventas.idxmin()

    destacados.append({
        'Tienda': nombre,
        'Producto Más Vendido': mas_vendido,
        'Cantidad Más Vendida': ventas.max(),
        'Producto Menos Vendido': menos_vendido,
        'Cantidad Menos Vendida': ventas.min()
    })

df_destacados = pd.DataFrame(destacados)

print("Productos más y menos vendidos por tienda:\n")
print(df_destacados)
```

## Desarrollo

En el grafico se observa que cada tienda tiene distintos articulos con mayor o menor venta.

Existe una dispersion de articulos mas y menos vendidos que no entregan mucha informacion para efectuar una recomendacion final.

Por ejemplo, en la tienda 1 su mayor venta fue microondas y tuvo menos ventas en los auriculares con microfono. Así, la tienda 4 vendió mas camas box y tuvo menos cantidad de guitarras electricas vendidas.

En cuanto a las cantidades vendidas, existe una similitud tanto en los mas vendidos, como en los menos vendidos. Siendo entre 57 y 65 articulos mas vendidos en las cuatro tiendas y los menos vendidos se encuentran entre 32 y 35 unidades, en las mismas cuatro tiendas

# 5. Envío promedio por tienda

```python
urls = [url1, url2, url3, url4]

tiendas = [pd.read_csv(url) for url in urls]
tienda1 = tiendas[0]
tienda2 = tiendas[1]
tienda3 = tiendas[2]
tienda4 = tiendas[3]

costo_envio_tienda1 = tienda1['Costo de envío'].mean()
costo_envio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_tienda4 = tienda4['Costo de envío'].mean()

print(f"El costo de envío promedio de la tienda 1 es: {costo_envio_tienda1:.2f}")
print(f"El costo de envío promedio de la tienda 2 es: {costo_envio_tienda2:.2f}")
print(f"El costo de envío promedio de la tienda 3 es: {costo_envio_tienda3:.2f}")
print(f"El costo de envío promedio de la tienda 4 es: {costo_envio_tienda4:.2f}")
```

## Desarrollo

Analizando el costo de envio por cada tienda, nos encontramos que la tienda 4 tiene un menor costo de envio con 23.459 versus la tienda 1 con 26.019.
